````python
#Define function to remove text beteen parenthesis
def remove_text_between_parentheses(text):
    """Removes text between parentheses using regular expressions."""
    if isinstance(text, str):
        return re.sub(r'\([^()]*\)', '', 'no city', text)
    else:
        return text
    return text
````
